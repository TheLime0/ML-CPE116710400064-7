# 2.  Fiber

## The recommended daily fiber intake is about 25 grams for adult women and 38 grams for adult men, according to general dietary guidelines.

## For older adults, the recommendations are slightly lower, around 21 grams for women and 30 grams for men. Adequate fiber intake supports digestive health, helps lower cholesterol levels, and aids in maintaining stable blood sugar levels.

```python
nutrition=['Data.Carbohydrate', 'Data.Cholesterol', 'Data.Fiber','Data.Protein', 'Data.Sugar Total', 'Data.Water','Data.Fat.Saturated Fat', 'Data.Fat.Total Lipid',
       'Data.Major Minerals.Calcium','Data.Major Minerals.Iron', 'Data.Major Minerals.Magnesium',
       'Data.Major Minerals.Phosphorus', 'Data.Major Minerals.Potassium',
       'Data.Major Minerals.Sodium', 'Data.Major Minerals.Zinc',
       'Data.Vitamins.Vitamin A - RAE', 'Data.Vitamins.Vitamin B12',
       'Data.Vitamins.Vitamin B6','Data.Vitamins.Vitamin C',
       'Data.Vitamins.Vitamin E']
```

```python
nutrition=['Data.Carbohydrate', 'Data.Fiber','Data.Protein', 'Data.Sugar Total', 'Data.Water']
for i in nutrition:
    df_fib=df.groupby(by='Category')[i].mean().sort_values(ascending=True).to_frame().reset_index().tail(25)
    plt.figure(figsize=(16,8))
    plt.title(f'Top Foods for High {i}', fontsize=16, pad=20.0)
    plt.xlabel(f'{i}', fontsize=15) 
    plt.ylabel('Foods', fontsize=18)  # The y-axis should be 'Counts' if it represents numerical values

    cmap = plt.get_cmap('RdBu')  # You can choose different colormaps like 'plasma', 'inferno', etc.
    colors = cmap(np.linspace(0, 1, len(df_fib[i])))



    plt.barh(y=df_fib['Category'], width=df_fib[i], color=colors, edgecolor='black', alpha=0.8)

# Display the plot
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.tight_layout()  # Adjust layout to fit labels
    plt.show()
    
```

```python
Fats=[ 'Data.Fat.Monosaturated Fat', 'Data.Fat.Polysaturated Fat','Data.Fat.Saturated Fat', 'Data.Fat.Total Lipid']
for i in Fats:
    df_fib=df.groupby(by='Category')[i].mean().sort_values(ascending=False).to_frame().reset_index().head(25)
    plt.figure(figsize=(16,8))
    plt.title(f'Top Foods for High {i}', fontsize=16, pad=20.0)
    plt.xlabel(f'{i}', fontsize=15) 
    plt.ylabel('Foods', fontsize=18)  # The y-axis should be 'Counts' if it represents numerical values
     
    cmap = plt.get_cmap('cividis')  # You can choose different colormaps like 'plasma', 'inferno', etc.
    colors = cmap(np.linspace(0, 1, len(df_fib[i])))


    plt.bar(x=df_fib['Category'],height=df_fib[i], color=colors, edgecolor='black', alpha=0.8)
    #plt.barh(y=df_fib['Category'], width=df_fib[i], color=colors, edgecolor='black', alpha=0.8)

# Display the plot
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.tight_layout()  # Adjust layout to fit labels
    plt.show()
```

```python
minerals=[ 'Data.Major Minerals.Calcium', 'Data.Major Minerals.Copper',
       'Data.Major Minerals.Iron', 'Data.Major Minerals.Magnesium',
       'Data.Major Minerals.Phosphorus', 'Data.Major Minerals.Potassium',
       'Data.Major Minerals.Sodium', 'Data.Major Minerals.Zinc']
for i in minerals:
    df_fib=df.groupby(by='Category')[i].mean().sort_values(ascending=True).to_frame().reset_index().tail(25)
    plt.figure(figsize=(16,8))
    plt.title(f'Top Foods for High {i}', fontsize=16, pad=20.0)
    plt.xlabel(f'{i}', fontsize=15) 
    plt.ylabel('Foods', fontsize=18)  # The y-axis should be 'Counts' if it represents numerical values
     
    cmap = plt.get_cmap('viridis')  # You can choose different colormaps like 'plasma', 'inferno', etc.
    colors = cmap(np.linspace(0, 1, len(df_fib[i])))


    #plt.bar(x=df_fib['Category'],height=df_fib[i], color=colors, edgecolor='black', alpha=0.8)
    plt.barh(y=df_fib['Category'], width=df_fib[i], color=colors, edgecolor='black', alpha=0.8)

# Display the plot
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.tight_layout()  # Adjust layout to fit labels
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
correlation_matrix = df_cor.corr()

# Set the size of the heatmap
plt.figure(figsize=(22, 26))

# Plot the heatmap
sns.heatmap(correlation_matrix[(correlation_matrix.abs()>0.7) & (correlation_matrix != 1)], annot=True, fmt='.2f', cmap='coolwarm', center=0, linewidths=0.5)

# Add title
plt.title('Nutritional Components Correlation Matrix', fontsize=16)

# Show the plot
plt.show()
```
